```python
import numpy as np
```

```python
#ques 1
arr=np.random.rand(10,1)
print("array:",arr)
m=np.mean(arr)
print("mean is:",m)
```

```python
# ques 2
arr=np.random.rand(20,1)
print("Array is:",arr)
variance=arr.var()
print("variance is:",variance)
st=arr.std()
print("Standard Deviation is: ",st)
```

```python
#ques 3
A=np.random.rand(10,20)
print(A)
B=np.random.rand(20,25)
print(B)

C=np.matmul(A,B)
print("multiply matrix",C)
print("sum of all elements",np.sum(C))
```
