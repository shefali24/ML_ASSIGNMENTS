# <h1> LINEAR REGRESSION</h1>

```python
df2=df[['quality','good','alcohol']]
```

```python
df2.head()
```

```python
X=df2[['good','alcohol']]
y=df2['quality']
```

```python
from sklearn.feature_selection import VarianceThreshold
z=VarianceThreshold(0.5).fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, \
                                                    random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# <h3>Making predictions</h3>

```python
y_pred = regressor.predict(X_test)
```

```python
dfp = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
dfp
```

```python
from sklearn import metrics

print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_pred))

print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_pred))

print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
print(metrics.r2_score(y_test, y_pred))
```

# <h3>Logistic Regression</h3> 

```python
X=df[['good','alcohol','chlorides','density','volatile acidity','residual sugar','free sulfur dioxide','pH','sulphates','total sulfur dioxide']]
y=df['quality']
```

# <h3> Test Train Split</h3>

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, \
                                                   y, \
                                                   test_size = 0.25, \
                                                   random_state = 0)
```

# <h3>Training</h3>

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

# <h3> Prediction</h3>

```python
y_pred = classifier.predict(X_test)
y_pred
```

# <h3> Measuring performance</h3>

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
print(metrics.classification_report(y_test, y_pred))
```
